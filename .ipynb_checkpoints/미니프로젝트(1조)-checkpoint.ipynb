```python
import numpy as np
import pandas as pd

import requests as req
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import matplotlib.pyplot as plt

import time

from tqdm import tqdm 
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
!pip install xlrd
!pip install openpyxl
!pip install pandas
```

```python
xlsx1 = pd.read_excel("./Report2023.xls")
xlsx1.to_csv("./Report.xls2023.csv")
```

```python
# 필요한 데이터만 전처리
a_f_2023 = pd.read_csv('./Report.xls2023.csv', skiprows=3)
a_f_2023 = a_f_2023[:-1]
#df = pd.read_csv('example.csv', skiprows=3)

accident_frequency2023 = a_f_2023[['광역지자체','사고[건]']]

accident_frequency2023.info()

p_s_2023 = pd.read_csv('./2023인구통계.csv', encoding='EUC-KR')
p_s_2023
population_density = p_s_2023[['행정구역','2023년12월_총인구수']]
population_density 
```

```python
population_density2023 = population_density.iloc[[i for i in range(len(population_density)) if i not in [0, 10]]]

population_density2023 = population_density2023.reset_index(drop=True)

result = pd.concat([accident_frequency2023, population_density2023], axis=1)

result
```

```python
# '인구수' 값에서 쉼표 제거 후 숫자형으로 변환
result['2023년12월_총인구수'] = result['2023년12월_총인구수'].apply(str).str.replace(',', '').astype(int)

# 그래프 그리기
fig, ax1 = plt.subplots(figsize=(12, 8))

# 첫 번째 y축: 사고 건수 (선 그래프)
ax1.plot(result['광역지자체'], result['사고[건]'], color='blue', alpha=0.8, marker='o', linestyle='--')
ax1.set_xlabel('광역지자체', fontsize=12)
ax1.set_ylabel('사고 건수', fontsize=12, color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# 두 번째 y축: 인구수 (막대 그래프)
ax2 = ax1.twinx()  # 두 번째 y축 생성
ax2.bar(result['광역지자체'], result['2023년12월_총인구수'], color='green', alpha=0.6)

ax2.set_ylabel('인구수', fontsize=12, color='green')
ax2.tick_params(axis='y', labelcolor='green')

# 두 번째 y축 범위 조정 (여유를 두어 범위 설정)
ax2.set_ylim([0, 15000000])

# 제목 및 x축 설정
plt.title('2023행정구역별 사고 건수 및 인구수', fontsize=14)
plt.xticks(rotation=45, fontsize=2) # x축의 라벨이 45도 회전하고, 글자 크기는 12로 설정

# 그래프 출력

plt.grid()
plt.show()
```
