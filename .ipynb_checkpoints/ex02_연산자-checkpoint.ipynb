### 연산자
 - 산술 연산자
      - 주의사항
      - / : 나누기(소수점까지 전부 표시, 실수형)
      - // : 몫(정수형)
      - % :  나머지

```python
num1 = 10
str2 = "7"

print(str(num1) + str2)
print(num1 + int(str2))
```

```python
num1 = 10
num2 = 7
print(num1 / num2)
print(num1 // num2)
print(num1 % num2)
```

```python
# 문자열 더하기
str1 = "안녕"
str2 = "하세요"

print(str1 + str2)
```

```python
# 문자열 곱하기
print(str1 * 10)
```

```python
str3 = "10"
num3 = 7

print(str3+ num3)

# 문자열과 숫자는 더 할 수 없다
```

```python
# 형변환을 통해서 자료형을 통일시켜야 함
# 문자열로 통일
print(str3+ str(num3))
# 정수형태로 통일
print(num3 + int(str3))
```

### 지수연산자
    사용방법 : **

```python
num = int(input("정수를 입력하세요 >> "))
power = int(input("지수를 입력하세요 >> "))

print(f'{num}의 {power}승은 {num**power}입니다.')
print('{}의 {}승은 {}입니다.'.format(num,power,num**power))
print('%d의 %d승은 %d입니다.' %(num,power,num**power))
```

### 대입(복합)연산자
    - 기존의 데이터에 더하기, 빼기, 나누기, 곱하기 등의 연산을 진행 후 다시 대입

```python
# 더하면서 대입

num = 5
print(num)
num += 3
print(num)
```

```python
# 빼면서 대입

num = 500
print(num)
num -= 450
print(num)
```

```python
# 곱하면서 대입

num = 3
print(num)
num *= 3
print(num)
```

```python
a = 3
b = 7

k = b # k = 7
b = a # b = 3
a = k # a = 7

print(a)
print(b)

```

```python
a,b = 3,7

a,b = b,a


print(a)
print(b)
```

### 비교연산자

    - a > b : a가 b보다 큰지 비교
    - a >= b : a가 b보다 크거나 같은지 비교
    - a < b : a가 b보다 작은지 비교
    - a <= b : a가 b보다 작거나 같은지 비교
    - a == b : a와 b가 같은지 비교
    - a != b : a와 b가 다른지 비교


```python
a = 10
b = 13

print(a > b)
print(a < b)
```

```python
import random

a =  random.randint(1,10) # 1~10까지 정수중 랜덤으로 하나를 가져오기
b =  random.randint(1,10)


print("a:{}, b:{}".format(a,b))
```

### 논리 연산자
    True, False의 논리값(boolan값)을 반환하는 연산자
    
    1. not 연산자
        논리를 뒤집는 역할

#### 멤버연산자

    in : 착고자하는 값이 포함되어 있느면 True
    not in : 찾고자 하는 값이 포함되어 있지 않으면 True    

```python
str1  = "파이썬 최고"

"파이썬" in str1

"파이썬" not in str1
```

### 삼항연산자

    a if 조건식 else b

```python
score = 80
"합격" if score >= 60 else "불합격"
```
