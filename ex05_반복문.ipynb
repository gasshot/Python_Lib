```python
current = int(input('현재 몸무게 >> '))
goal = int(input('목표 몸무게 >> '))
week = 1


while True :
    low = int(input(f'{week}주차 감량 몸무게 : '))
    current -= low # 복합대입연산자
    week += 1
    
    if current =< goal:
        print(f'{current} kg 달성!! 축하합니다!')
        break
```

```python
# 숫자를 입력 받고 입력 받은 숫자의 약수를 구하시오
# (약수란 어떤 수를 나누어 떨어지게 하는 수)

num7 = int(input("정수 입력 >> "))
str2 = '' 


for i in range(1,num7 + 1):
    if num7 % i == 0:
        str2 += str(i) + " "

print(f'{num7}의 약수 : ' + str2)
```

```python
# 시작값과 종료값을 입력받고 각 숫자의 약수의 합을 구하시오

startNum = int(input("Start >> "))
endNum = int(input("End >> "))


for i in range(startNum, endNum +1):
    total = 0
    
    for j in range(1,i + 1):
        if i % j == 0:        
            total += j
    
    print (f'{i}의 약수의 합 : {total} ')
```
