### 함수의 특성
    - 매개변수의 순서대로 인자가 들어감

    - 기본값을 설정해줄 수 있음
    - 리턴의 결과값은 하나, 두개 이상 적으면 패킹되어서 튜플로 나옴
    - 입력값이나 결과값이 없을 수도 있음

```python
def number_subtraction (num1, num2):

    result = int(num1) - int(num2)
    return result
```

```python
def cal(num1,num2,str1):
    '''덧셈과 뺄셈, 곱셈을 계산하는 함수'''
    if(str1 == '+'):
        result = num1 + num2
    elif(str1 == '-'):
        result = num1 - num2
    elif(str1 == '*'):
        result = num1 * num2
    return result   
        
```

```python
# 두수를 입력 받아서 원하는 연산을 수행하여 결과를  리턴하는 함수를 정희하시오

num1 = int(input("첫 번째 정수 입력 >> "))
num2 = int(input("두 번째 정수 입력 >> "))

while True:
    op = input("연산자 입력(+, -) >> ")
    if(op == '+' or op == '-'):
        break


result = cal(num1,num2,op)

print(f'결과 : {result}')



```

```python
# 가변 매개변수(variable parameter)를 활용해서 모든 숫자를 더해서 반환하는 함수를 정의

def add(*args):
    print(args) # 전달된 모든 인수(매개변수, 값)는 튜플의 형태로 저장
    return sum(args) # args의 값들이 튜플의 형태로 저장이 되기 때문에 사용 가능
```

```python
add(1,2,3)
```

```python
def add1(n, *args):
    print(args)
    return sum(args)

def add2(*args, n):
    print(args)
    return sum(args)
```

```python
add1(1,2,3)

add2(1,2,3 , n = 1) # 값을 직업 지정해서 넣어주면 가능
# 가변매개변수 agrs가 모든 입력값을 패킹하기 때문에 n의 값을 넣을 수 없음
```

```python
# 키워드 가변매개변수
# 함수 호출 시 몇 개의 인수가 전달될 지 알 수 없으면서
# 딕셔너리 형태로 패킹하여 사용하고 싶을 경우 선언(**kwargs)


def print_map(**kwargs):
    print(kwargs)
```

```python
print_map(a = 1, b = 2, c = 100)
```

- 함수 정의 시 매개변수 순서
1. 일반 매개변수 : num1, num2
2. 가변 매개변수 : *args
3. 키워드 매개변수 : power = 2 기본값 설정
4. 키워드 가변 매개변수 : **kwargs
