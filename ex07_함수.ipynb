```python
def cal(num1,num2,str1):
    ''' ,   '''
    if(str1 == '+'):
        result = num1 + num2
    elif(str1 == '-'):
        result = num1 - num2
    elif(str1 == '*'):
        result = num1 * num2
    return result   
        
```

```python
#           

num1 = int(input("    >> "))
num2 = int(input("    >> "))

while True:
    op = input(" (+, -) >> ")
    if(op == '+' or op == '-'):
        break


result = cal(num1,num2,op)

print(f' : {result}')



```
