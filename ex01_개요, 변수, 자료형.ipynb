### 문자열 인덱싱과 슬라이싱
- 문자열은 처음부터 끝까지 인데스 번호를 지정받는다
    - 주의1. 인덱스 번호는 0번부터 시작한다
    - 주의2. 띄어쓰기도 인덱스 번호를 지정받는다
- 인덱싱 : 문자열 안에서 내가 원하는 데이터 하나를 뽑아올 때 사용
    - 인덱싱 방법 : 변수명[원하는 인덱스 번호]
- 슬라이싱 : 문자열 안에서 범위를 주고 잘올 때 사용
    - 슬라이싱 방법: 변수명[시작 인덱스번호 : 끝 인덱스 번호]
    - (끝 인덱스 번호 - 1)인덱스 번호까지 문자열을 가져온다

```python
s ='My name is YH'


print(s[5])
print(s[0:6])
print(len(s))


#대문자 Y(이름 앞글자) 가져오기
print(s[-2])
```

```python
# 슬라이싱 : 변수명[시작인덱스번호 : 끝인덱스번호 + 1] 콜(:)론으로
# My 가져오기
s ='My name is YH'
print(s[0:2])
print(s[11:13])
print(s[-2:0])
print(s[-2:])
# 콜론 뒤를 비워두면 "끝까지" 가져오라는 의미


# 파이썬의 슬라이싱은 항상 시작 인덱스(start)가 끝 인덱스(end)보다 작아야 하고, 진행 방향은 왼쪽에서 오른쪽입니다.
# 하지만 -2는 문자열의 끝쪽에 위치하고, 0은 문자열의 맨 앞쪽에 있으므로, 왼쪽에서 오른쪽으로 진행이 불가능합니다.
```

```python
day = '2024년 12월 13일의 날씨는 흐림입니다.'


date = day[0:13]
weather = day[19:21]


print('날짜 :', date)
print('날짜 : {}'.format(date))
print(f'날짜 : {date}')
print('날씨 :', weather)
```

```python
# 해당 문자 위치 알려주는 함 :  변수명.find('문자'), 변수명.index('문자')

day.index('일')
day.find('흐')
print(day[:13])
print(day.find('흐'))

# 없는 문자 검색하기
day.find('뷁')
# find()를 사용해서 없는 문자를 검색하면 -1 반환
```

```python
# index()를 사용해서 없는 무자를 검색하면 오류가 발생
day.index('뷁')
```

```python
# 문자열 길이 반환 :  len()
len(day)
```

```python
# 문자열에 포함된 문자 개수 세기 : count()
s = 'python is very easy.'

# s라는 변수에 e문자가 몇번 들었나요?

s.count('e')
```

### 문자열 포매팅
    - 문자열 안에 특정한 값을 사용해서 편리하게 수정할 수 있다.
    - %기호 포매팅
    - fomat 함수 포매팅
    - 문자열 포매팅
    


```python
monthValue = 7
dayValue = 7
strings = '오늘은 %d월 %d일 입니다.'%(monthValue, dayValue)

# %기호 포매팅


strings
```

```python
# fomat 함수 포매팅
monthValue = 12
dayValue = 13

strings1 = '오늘은 {}월 {}일 입니다.'.format(monthValue, dayValue)
strings1
```

```python
# f문자열 포매팅

monthValue = 12
dayValue = 14

strings2 = f'오늘은 {monthValue}월 {dayValue}일 입니다.'
strings2
```

```python
x = 100
y = 200

sum2 = x + y

strings3 = f'{x}와 {y}의 합은 {sum2}입니다.'
strings4 = '{}와 {}의 합은 {}입니다.'.format(x,y,sum2)
strings5 = '%d와 %d의 합은 %d입니다.'%(x,y,sum2)
strings6 = f'{x}와 {y}의 합은 {x+y}입니다.'

print(strings3)
print(strings4)
print(strings5)
print(strings6)
```
