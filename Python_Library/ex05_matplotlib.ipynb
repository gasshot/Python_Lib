```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
y = [2,4,6]
y

# plot() : 

plt.plot(y)

# 시각화
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]

# x축과 y축을 설정하고 그래프를 그리기
plt.plot(x,y)

# 시각화
plt.show()
```

```python
plt.plot(x,y ,linestyle = ':')

# 시각화
plt.show()
```

```python
plt.plot(x,y ,marker = '*')
```

```python
x = [0,1,2,3,4,5,6,7]
y = [5,2,3,6,8,9,5,7]
plt.plot(x,y ,color = 'g' ,linestyle = '--',marker = '*')
```

```python
# grid : 틱의 위치를 잘 보기 위해서 설정
plt.plot(x,y ,color = 'g' ,linestyle = '--',marker = '*')

plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.grid()
plt.show()
```

```python
x = [1,2,3,4]
z = [2,4,6,8]
y = [3,6,9,12]

plt.plot(x)
plt.plot(y)
plt.plot(z)

plt.grid()
plt.show()
```

```python
plt.rc('font', family = 'Malgun Gothic')
```

```python
#범례 표시하기

#범례설정
plt.plot(x, label = '엑스')
plt.plot(y, label = '와이')
plt.plot(z, label = '제트')

# 범례 표시
plt.legend()
plt.show()
```
