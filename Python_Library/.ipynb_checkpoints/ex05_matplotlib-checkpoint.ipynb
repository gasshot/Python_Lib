```python
hr = pd.read_csv('./data/hr-info.csv',encoding='EUC-KR',parse_dates =['HIRE_DATE'] )
hr.index = range(1,len(hr.index) + 1)
hr['HIRE_YEAR'] = hr['HIRE_DATE'].dt.year


x_array = hr['HIRE_YEAR'].unique()
# hr.groupby(by = 'HIRE_YEAR').count()

y_array = hr['HIRE_YEAR'].value_counts().sort_index().plot(kind = 'line',
                                                          xlabel = "HIRE_DATE",
                                                          ylabel = 'HIRE_COUNT',
                                                          grid = True)

plt.show()

```

```python
x_edu_level = hr['EDU_LEVEL'].unique()
x_edu_level.sort()
my_list = {}
for item in hr['EDU_LEVEL'].unique():
    # print(hr[hr['EDU_LEVEL'] == str(item)]['SALARY'].mean())
    my_list[str(item)] = hr[hr['EDU_LEVEL'] == str(item)]['SALARY'].mean()


print(my_list)
# 딕셔너리를 Pandas Series로 변환
my_series = pd.Series(my_list)

my_series.plot(kind = 'bar',)
```

```python
# 1) 학위, 성별을 고려한 평균 급여를 추출
edu_gender_mean = hr.groupby(by = ['EDU_LEVEL', 'GENDER'])['SALARY'].mean()
# 2) 시작화 > bar
edu_gender_mean.plot(kind = 'bar', colormap = 'Set3', rot = 70)

# 범례 표시
plt.legend()  # 범례와 제목 설정
```

```python
# 데이터를 재 구조화.
# pivot(피벗)하여 새로운 데이터프레임 생성

sal_mean = edu_gender_mean.reset_index().pivot(index = 'EDU_LEVEL', columns = 'GENDER', values = 'SALARY')

sal_mean.plot(kind = 'bar',
             colormap = 'Set3',
              xlabel = '')
```
