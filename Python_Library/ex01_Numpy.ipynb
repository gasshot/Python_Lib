### Indexing & Slicing
- Indexing (색인, 가져오다) : 원하는 요소에 접근하여 색인(인덱스)를 통해서 가져오다.
- Slicing(잘라내어 가져오다) : 원하는 요소에 접근하여 범위를 가져오다.
- (시작 인덱스, 끝 인덱스) -> 끝 인데스는 포함이 되지 않습니다.(파이썬의 기본 슬라이싱 방식에 따르는 표준이다.)


```python
# 실습) 1~ 70 까지의 범위의 숫자를 가진 1차원 배열을 생성해서 my_arr1변수에 저장

my_arr1 = np.arange(1,71)
my_arr1

my_arr1.shape

print(my_arr1)
# 실습) (7,10) 2차원 형태의 배열로 reshape를 저장

my_arr1 = my_arr1.reshape((7,10))
my_arr1
```

```python
# array명[행, 열] 튜플 인덱싱이라고 부른다
# , 를 기준으로 앞에는 행, 뒤는 열 데이터에 접근 가능하다.


int_array = [my_arr1[3,4],my_arr1[5,6],my_arr1[2,1]]

# 35만 출력을 해보세요
my_arr1[3,4]
# 57만 출력을 해보세요
my_arr1[5,6]
# 22만 출력을 해보세요
my_arr1[2,1]

for i in int_array:
    print(i)
```

```python
# 슬라이싱
# array명[행 시작값 : 행 끝값, 열 시작값 열 끝값]

# 24,34,44 데이터만 가져오기
my_arr1[2 : 5, 3]
my_arr1[2 : 5, 3 : 4]

# 46 47 56 57

my_arr1[3 : 5, 5: 7]

# 실습1) 34~39, 44~49

print(my_arr1[3 : 5, 3: 9])

# 실습2) 12~19,22~29,32~39,42~49,52~59
print(my_arr1[1 : 6, 1 : 9])
```

### Boolean 색인
- 원하는 요소를 Boolean type(True or False)로 특정 조건에 따라 "요소"를 가져오는 것
- Numpy의 최고 장점

```python
score = np.array([80,90,75,77,60])
print(score)

# 80점 이상인 점수만 가져오기
score[score >= 80]

score
# 80점 이하인 데이터만 불리언 색인

score[score <= 80]
```
