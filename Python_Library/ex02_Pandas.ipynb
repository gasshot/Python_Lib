```python
p_2020 = pd.Series([9602000,3344000,1488000,2419000],index = [ '서울','부산','광주','대구'])
p_2020.name = '인구수'
p_2020.index.name = '도시'

p_2020
```

### Series 색인
    - 인덱싱
    - 슬라이싱
    - 불리언

```python
# 인덱싱
# 서울 인덱스의 values 출력

p_2015['서울']

p_2015[['서울', '인천']]

# 대괄호를 한번 더 씌어주면 리스트 형태로 데이터를 추출할 수가 있다.
p_2015[['서울', '인천', '서울']]
```

```python
# 슬라이싱
# [시작 값 : 끝 값]

p_2015['부산' : '서울'] # 행 인덱스로 접근을 하는 것이 아니라 Location 위치로 바로 접근을 하기 때문에
```

```python
p_2015[1:3]
```

```python
# 불리언 색인
p_2020[p_2020 > 3000000]

# 논리연산자
# AND -> &
# OR -> ||

# 인구수가 300만 이상이면서 500만 이하인 도시만 조회

print((p_2020 >= 3000000) & (p_2020 <= 5000000))
```

```python
# notnull() , isnull()
# notnull() : null이 아닌 값을 조회.
# isnull() : null인 값을 조회.




p_null = p_2015 + p_2020

p_null
```

```python
p_null[p_null.isnull()]

p_null[p_null.notnull()]
```

```python
# print(p_2015)
# print(p_2020)

result =(( p_2020 - p_2015 ) / (p_2015)) * 100
check_array =(( p_2020 - p_2015 ) / (p_2015)) * 100

round(check_array[check_array.notnull()],ndigits = 2)
```

```python
# 데이터를 갱신, 추가, 삭제

# 데이터 갱신
# null값을 채워주자

result['광주'] = 1.5

result['목포'] = 1.8

result
```

```python
# 데이터 삭제
# del

del result['목포']

result
```

### DataFrame 사용하기
- 2차원 행렬 데이터를 표(엑셀) 형태로 저장
- 가로인 행(row) 과 세로인 열(column)으로 이루어졌다.

```python
score_list = np.array([[90,100,80],[100,80,75],[95,80,100]])
score_list.shape

score_df = pd.DataFrame([[90,100,80],[100,80,75],[95,80,100]], columns = ['A','B','C'], index = ['DB','Python','Java'])

score_df
```

```python
score_df2 = pd.DataFrame({'A' : [90,100,95] ,'B' : [100,80,80] ,'C' : [80,75,100]},index = ['DB','Python','Java'])

score_df2
```

```python
# 행 / 열 변경

score_df = score_df.T

score_df
```

### DataFrame 정보를 확인하기
- df명.values : 값을 확인
- df명.index : 인덱스를 확인
- df명.columns : 컬럼값을 학인

```python
# score_df 의 값 확인
score_df.values
# score_df 의 인덱스 확인
score_df.index
# score_df 의 컬럼 값 확인
score_df.columns
```

```python
# 컬럼 추가하기

score_df['ML'] = [90,80,100]

score_df
```

```python
# 특정컬럼 삭제하기

del score_df['ML']


score_df
```

### DataFrame 색인
- 컬럼 단위로 데이터 추출
    - df명['컬럼이름']
- 행 단위로 데이터 추출
    - loc(location) :  인덱스를 기준으로 행을추출, 커럼 이름으로 접근
    - iloc(integer location) : 행 번호를 기준으로 데이터를 추출, 행 번호로 접근
    - ex) df명.loc[행, 열] --> ,(콤마)를 기준으로 행, 열(앞이 행 뒤열)

```python
score_df
```

```python
# 컬럼 단위로 데이터를 추출

# 'ML'컬럼 의 데이터를 추출
score_df['ML']
```

```python
# 슬라이싱
score_df.loc['B',"Java"]


# 실습 C행의 ML점수를 출력 > 100

score_df.loc['C','ML']

# 실습 A행의 Java점수를 출력 > 95

score_df.loc['A','Java']
```

```python
score_df.loc['A' : 'B','Python' : 'Java']


score_df.loc['B' : 'C','Java']
```

```python
# iloc(integer location) : 행 번호 위치로 접근하겠다 > 넘파이스럽게 접근하는 것
# iloc[행, 열]
# iloc[행 시작값 : 행 끝값, 열 시작값 : 열 끝값] -> 끝밧은 포함이 되지 않아서 +1을 해줘야한다.

print(score_df)
score_df.iloc[0:2,0:2]
```
