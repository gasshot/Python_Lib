# 2. 데이터 수집
- 정의된 문제를 해결하기위한 데이터 모으기

```python
import numpy as np
import pandas as pd
# 수집된 데이터를 numpy or pandas를 사용해서 read하기
# pd.read_csv('경로')
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

# 3. 데이터 전처리
- 수집된 데이터를 학습할 수 있는 형태로 처리하기

```python
# value_counts() : 컬럼안에서 값의 갯수 세기
# fillna() : 결측치 채우기
# info() : 데이터 전반적으로 확인하기
```

```python
# X와 y나누기
X = cancer['data']
y = cancer['target']
```

```python
# train과 test구분하기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3,
                                                   random_state = 25)
```

# 4. 탐색적 데이터 분석
- 데이터의 특징, 생김새 확인하기

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# 5. 모델 선택 및 하이퍼 파라미터 튜닝
- 데이터에 어울리는 모델 찾기
- 모델 세부 기능 설정하기

```python
# 머신러닝에서는 sklearn
# 딥러닝에서는 tensorflow
```

```python
from sklearn.ensemble import RandomForestClassifier , AdaBoostClassifier
# RandomForest : Bagging 계열의 모델, 병렬 구조
# AdaBoost : Boosting 계열의 모델, 순차 구조
# 여러개의 DecisionTree를 사용한 모델
# Bagging : 하이퍼 파라미터가 DT의 파라미터 + a
# n_estimators = 사용할 DT의 갯수
# max_features = 데이터 분할할 때 사용할 특성의 비율
rf = RandomForestClassifier(max_depth = 3, n_estimators = 75, max_features=0.5)
# Boosting : DT 파라미터 X
ab = AdaBoostClassifier(n_estimators = 75)
```

# 6. 학습
- 모델이 데이터에서 규칙 찾기

```python
# model.fit(X_train, y_train)
```

```python
rf.fit(X_train,y_train)
ab.fit(X_train,y_train)
```

# 7. 평가 및 예측
- 평가 : 만들어진 규칙이 얼마나 정확한지 알아보기
- 예측 : 만들어진 규칙 사용해보기

```python
# model.predict(X_test)
# model.score(X_test, y_test)
```

```python
print(rf.score(X_train,y_train))
print(rf.score(X_test,y_test))
print('*'*20)
print(ab.score(X_train,y_train))
print(ab.score(X_test,y_test))
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import cross_val_score
rf_cv = cross_val_score(rf, X_train,y_train, cv = 5)
ab_cv = cross_val_score(ab, X_train,y_train, cv = 5)
print(rf_cv)
print(rf_cv.mean())
print('*'*20)
print(ab_cv)
print(ab_cv.mean())
```

```python
from sklearn.metrics import classification_report
y_pred = rf.predict(X_test)
print(classification_report(y_test,y_pred))
# precision - recall이 0.95 - 0.93에서 0.98 - 0.91로 바뀜 > 편차가 커짐 > 안좋아짐
```

```python
from sklearn.metrics import classification_report
y_pred = ab.predict(X_test)
print(classification_report(y_test,y_pred))
```
