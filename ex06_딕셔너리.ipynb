# 딕셔너리


    
    


- 딕셔너리명 = {Key : Value, Key : Value, Key : Value,....}
- Key는 변하지 않는 값을 사용 (중복이 안됨)
- Value는 변하는 값, 변하지 않는 값 모두 사용 가능(중복가능)

```python
a = {}

b= {'name' : 'yh','pn' : '010-1111-1111','bd' : '0111'}

c = {1:123, 2:'abc'}


print(b)
print(c)
```

```python
# key가 중복이 되면 안되는 이유
d = {'name' : 'yh','name' : 'dy'}

d
```

```python
# value는 중복값 가능
e = {'name' : 'yh','nickname' : 'yh'}
e
```

```python
# 딕셔너리 값 추가하기
# 딕셔너리명[새로운 key] = value

b['age'] = 20
b
```

```python
# 딕셔너리 값 수정
# 딕셔너리명[존재하는 key] = value

b['pn'] = '010-2222-2222'
b
```

```python
# dic_test 에 다음을 딕셔너리로 만들기

dic_test = {'노래제목' : 'SPOT!'}

# 딕셔너리 값 추가해서 다음의 정보를 저장 시키기

dic_test['가수'] = '지코'
dic_test['날짜'] = '2024.04.26'

dic_test
```

```python
# 딕셔너리 값 삭제
# del 딕셔너리명[지우고싶은 key]
del(b['age'])
b

```

```python
# 딕셔너리 값 모두 지우기
# 딕셔너리명.clear()
b.clear()
```

```python
# 딕셔너리 값 가져오기
b = {'name' : 'yh','pn' : '010-1111-1111','bd' : '0111'}
b['name']
```

```python
# 딕셔너리명.get(key)
b.get('name')
```

```python
# b['age']
b.get('age')

if(b.get('age') == None):
    print('없는 데이터')

```

```python
#딕셔너리 key만 가져오기
#딕셔너리명.keys()

b.keys()
```

```python
#딕셔너리 value만 가져오기
#딕셔너리명.values()
b.values()
```

```python
#딕셔너리 key, value만 가져오기
#딕셔너리명.items()
b.items()
```

```python
#리스트화


list77 = list(b.items())

list77
```

```python
# 리스트이므로 인덱싱, 슬라이싱 가능
list(b.items())[0]
```

```python
# 튜플이므로 인덱싱, 슬라이싱 가능
list(b.items())[0][0]
```

```python
# for문 활용
for key in list(b.keys()):
    print(key)
```
