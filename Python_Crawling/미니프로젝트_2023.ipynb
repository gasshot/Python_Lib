### 2023의 경우 메서드가 기존것과 차이가 있으니 참고

```python
# CSV 파일 읽기
population_report = pd.read_csv('./미니프로젝트/2023인구통계.csv', encoding='EUC-KR')

# 빈 딕셔너리 생성
population_result_2023 = {'행정구역': [], '인구수': []}

# 데이터 추출
for i in range(1, len(population_report)):  # len(report) -> len(report_2021)
    population_result_2023['행정구역'].append(population_report['행정구역'][i])
    population_result_2023['인구수'].append(population_report['2023년12월_총인구수'][i])

# DataFrame 생성
population_2023 = pd.DataFrame(population_result_2023)

# 결측치 및 문자열 처리
population_2023['행정구역'] = population_2023['행정구역'].fillna('').str.replace(r'\s*\(.*?\)', '', regex=True)

# '행정구역' 컬럼에서 값이 '강원도'인 행을 제외
population_2023 = population_2023[population_2023['행정구역'] != '강원도']

# 인덱스 재정렬 (선택 사항)
population_2023 = population_2023.reset_index(drop=True)

# '행정구역' 값이 '강원특별자치도'인 행을 '강원도'로 변경
population_2023.loc[population_2023['행정구역'] == '강원특별자치도', '행정구역'] = '강원도'

# 결과 출력
population_2023
```
