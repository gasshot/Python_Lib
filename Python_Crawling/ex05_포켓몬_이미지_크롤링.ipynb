```python
# 컴퓨터용 브라우저 실행

driver = wb.Chrome()

driver.get('https://www.pokemonkorea.co.kr/pokedex#pokedex_1')
```

```python
def check():
    result = driver.execute_script('return document.body.scrollHeight')
    return result
```

```python
# 스크롤 내리기 추가!
# 한솥 도시락 데이터 수집 실습에서는 횟수가 제한(더보기 버튼 개수가 제한)
# 여기서는 ? 페이지 길이 가져오기

print(check()) # 스크롤 길이 확인

# 창이 작으면 스크롤 해야할 페이지가 길어지니까 숫자가 더 큼
```

### 화면 스크롤 내리기
- selenium

```python
driver.page_source # 문자열로된 html 문서
```

```python
# END가 가장 효율적!
# Keys.END
import time as tm
body = driver.find_element(By.TAG_NAME, 'body')
```

```python
# 페이지 스크롤 알고리즘
while True :
    old = check()
    
    body = driver.find_element(By.TAG_NAME, 'body')
    body.send_keys(Keys.END)
    tm.sleep(1)

    new = check()
    if old == new:
        print('스크롤 하강 완료')
        break
    else :
        print('스크롤 진행 중...')
        pass
```

```python
imgBox = driver.find_elements(By.CSS_SELECTOR, 'img.img-fluid') 
# print(len(imgSrc))
imgSrc = []

for img in imgBox:
    imgSrc.append(img.get_attribute('src'))
    print(img.get_attribute('src'))
```
