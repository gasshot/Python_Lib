```python
# 운영체제를 다룰 수 있는 라이브러리
import os # 폴더 생성, 파일 삭제 등 운영체제에 접근해서 관리 할 수 있는 라이브러리

# 크롤링 라이브러리(Selenium)
from selenium import webdriver as wb # 컴퓨터용 브라우저
from selenium.webdriver.common.by import By # Selenium 라이브러리에서 웹 요소
from selenium.webdriver.common.keys import Keys # 키보드 이벤트를 시뮬레이션

import time

from tqdm import tqdm # 반복문 진행 현황 출력하는 도구
```

```python
# 웹 드라이버 설정
driver = wb.Chrome()
```

```python
url = 'https://map.naver.com/p?c=15.00,0,0,0,dh'
# 웹 페이지 열기
driver.get(url)
```

```python
# 검색 -> 검색란 가져오고, 검색어 입력
query = input('검색할 단어를 입력 >>')
# 검색창 찾기
search = driver.find_element(By.CSS_SELECTOR,'input.input_search') 

# 검색어 입력 후 엔터 키 시뮬레이션 + 엔터 키 누르기
search.send_keys(query + "\n") # '\n = 엔터와 같다'
```
