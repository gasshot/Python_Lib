```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
case_report = pd.read_csv('./Report.csv')
case_result_2021 = {'광역지자체': [] , '사고[건]' : []}


for i in range(len(case_report)):
    case_result_2021['광역지자체'].append(case_report['광역지자체'][i])
    case_result_2021['사고[건]'].append(case_report['사고[건]'][i])
    

case_2021 = pd.DataFrame(case_result_2021)
case_2021
```

```python
# DataFrame으로 변환
df = pd.DataFrame(case_2021)

# 사고[건] 데이터를 숫자형으로 변환
df['사고[건]'] = df['사고[건]'].str.replace(',', '').astype(int)

# 그래프 생성
plt.figure(figsize=(12, 6))
plt.plot(df['광역지자체'], df['사고[건]'], color='blue', alpha=0.8 , marker = 'o', linestyle = '--')
plt.xlabel('광역지자체', fontsize=12)
plt.ylabel('사고 건수', fontsize=12)
plt.title('광역지자체별 사고 건수', fontsize=14)
plt.xticks(rotation=0, fontsize=10)
plt.tight_layout()

# 그래프 출력
plt.show()
```

```python
# CSV 파일 읽기
population_2021 = pd.read_csv('./미니프로젝트/2021인구통계.csv', encoding='EUC-KR')

# 빈 딕셔너리 생성
population_result_2021 = {'행정구역': [], '인구수': []}

# 데이터 추출
for i in range(1, len(population_2021)):  # len(report) -> len(report_2021)
    population_2021['행정구역'].append(population_2021['행정구역'][i])
    population_2021['인구수'].append(population_2021['2021년12월_총인구수'][i])

# DataFrame 생성
xxxx = pd.DataFrame(population_2021)

# 결측치 및 문자열 처리
xxxx['행정구역'] = xxxx['행정구역'].fillna('').str.replace(r'\s*\(.*?\)', '', regex=True)

# 결과 출력
xxxx
```

```python
# 첫 번째 데이터셋 (사건수)
data1 = {
    '행정구역': ['서울', '부산', '대구', '인천', '광주', '대전', '울산', '세종', '경기', '강원', '충북', '충남', '전북', '전남', '경북', '경남', '제주'],
    '사건수': [33913, 11339, 12133, 7912, 7543, 7027, 3801, 788, 53332, 6605, 7985, 8423, 6112, 8685, 12100, 11059, 4373]
}

df1 = pd.DataFrame(data1)

# 두 번째 데이터셋 (인구수)
data2 = {
    '행정구역': ['서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시', '울산광역시', '세종특별자치시', '경기도', '강원도', '충청북도', '충청남도', '전라북도', '전라남도', '경상북도', '경상남도', '제주특별자치도'],
    '인구수': [9509458, 3350380, 2385412, 2948375, 1441611, 1452251, 1121592, 371895, 13565450, 1538492, 1597427, 2119257, 1786855, 1832803, 2626609, 3314183, 676759]
}

df2 = pd.DataFrame(data2)

# 행정구역 이름 변환 (첫 번째 데이터셋의 이름을 두 번째 데이터셋에 맞게 변환)
name_mapping = {
    '서울': '서울특별시',
    '부산': '부산광역시',
    '대구': '대구광역시',
    '인천': '인천광역시',
    '광주': '광주광역시',
    '대전': '대전광역시',
    '울산': '울산광역시',
    '세종': '세종특별자치시',
    '경기': '경기도',
    '강원': '강원도',
    '충북': '충청북도',
    '충남': '충청남도',
    '전북': '전라북도',
    '전남': '전라남도',
    '경북': '경상북도',
    '경남': '경상남도',
    '제주': '제주특별자치도'
}

# 변환된 이름을 첫 번째 데이터셋에 적용
df1['행정구역'] = df1['행정구역'].map(name_mapping)

# 두 데이터프레임을 '행정구역'을 기준으로 병합
merged_df = pd.merge(df1, df2, on='행정구역', how='inner', suffixes=('_사건수', '_인구수'))

# DataFrame을 다른 형태로 출력 (예시: 리스트로 변환)
result_list = merged_df.values.tolist()

# 결과 출력 (리스트 형식)
print(result_list)
```

```python
# 데이터 생성
x = np.linspace(0, 10, 100)
y1 = np.sin(x)  # 첫 번째 y축 데이터
y2 = np.cos(x)  # 두 번째 y축 데이터

# 기본 플롯 생성
fig, ax1 = plt.subplots()

# 첫 번째 y축 (ax1)
ax1.plot(x, y1, 'b-', label='Sine')  # 파란색 선
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Sine', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# 두 번째 y축 (ax2)
ax2 = ax1.twinx()  # x축을 공유하는 새로운 y축 생성
ax2.plot(x, y2, 'r--', label='Cosine')  # 빨간색 점선
ax2.set_ylabel('Cosine', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# 제목 설정
plt.title("Dual Y-axis Example")

# 범례 추가
fig.tight_layout()  # 레이아웃 조정
plt.show()
```

```python
# CSV 파일 읽기
report_2023 = pd.read_csv('./미니프로젝트/2023인구통계.csv', encoding='EUC-KR')

# 빈 딕셔너리 생성
population_2023 = {'행정구역': [], '인구수': []}

# 데이터 추출
for i in range(1, len(report_2023)):  # len(report) -> len(report_2021)
    population_2023['행정구역'].append(report_2023['행정구역'][i])
    population_2023['인구수'].append(report_2023['2023년12월_총인구수'][i])

# DataFrame 생성
xxxx = pd.DataFrame(population_2023)

# 결측치 및 문자열 처리
xxxx['행정구역'] = xxxx['행정구역'].fillna('').str.replace(r'\s*\(.*?\)', '', regex=True)

# '행정구역' 컬럼에서 값이 '강원도'인 행을 제외
xxxx = xxxx[xxxx['행정구역'] != '강원도']

# 인덱스 재정렬 (선택 사항)
xxxx = xxxx.reset_index(drop=True)

# '행정구역' 값이 '강원특별자치도'인 행을 '강원도'로 변경
xxxx.loc[xxxx['행정구역'] == '강원특별자치도', '행정구역'] = '강원도'

# 결과 출력
xxxx
```
